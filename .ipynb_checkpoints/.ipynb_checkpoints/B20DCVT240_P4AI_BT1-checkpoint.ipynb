###  Chuẩn hóa dữ liệu theo MinMax
- Xnew : Giá trị sau khi chuẩn hóa
- Xi : Giá trị trước khi chuẩn hóa
- min(X) : Giá trị nhỏ nhất của trường dữ liệu
- max(X) : Giá trị lớn nhất của trường dữ liệu

<img src = "min_max_normalisation.jpg" width='400'>
     

```python
df3 = df1.copy()
```

```python
df3.drop('variety', axis=1, inplace=True)
minX = df3.min()
maxX = df3.max()
```

```python
for i in df3.columns:
    k = 0
    df3[i] = (df3[i] - minX[k])/(maxX[k] - minX[k])
    k = k + 1
df3
```

# Liên tục hóa thuộc tính variety

# Sử dụng PCA và t-SNE
- PCA : linear Dimensionality reduction, find decide on how much variance
- t-SNE : non-linear Dimensionality reduction,can handle outliers

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
n_comp = 2
pca = PCA(n_components=n_comp, svd_solver='full', random_state=1001)
X_pca = pca.fit_transform(X)
for j in range(n_comp):
    print(pca.explained_variance_ratio_[j])
```

```python
tsne = TSNE(n_components = n_comp, init='pca', random_state=1001, perplexity=30, method='barnes_hut')
#tsne = TSNE(n_components=2, learning_rate='auto',init='random', perplexity=3)
X_tsne = tsne.fit_transform(X)
X_tsne.shape   
```

# Phân cụm dữ liệu sử dụng DBSCAN (Density-Based Spatial Clustering Of Applications With Noise)
- distance between nearest points

```python
from sklearn.cluster import DBSCAN
```

```python
X_tsne
```

```python
df7 = X_tsne.copy()
db = DBSCAN(eps=0.5,min_samples=5,metric='euclidean')
model=db.fit(df7)
label=model.labels_
label
```

```python
#loại bỏ những điểm ngoại lai -1
non_outlier_indices = np.where(label != -1)
X_non_outliers = label[non_outlier_indices]
X_non_outliers
```
